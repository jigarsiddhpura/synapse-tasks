```python
from google.colab import drive
drive.mount('/content/gdrive')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/content/gdrive/MyDrive/Task2hp/HPSpells.csv')
df
```

```python
df.head()
```

```python
df.drop(df[df.Type=='Hex'].index,inplace=True)
```

```python
df.drop(df[df.Light=='NaN'].index,inplace=True)
```

```python
freq = {}

for i in df.Type.unique():
  ctr = 0
  for j in df.Type:
    if i==j:
      ctr += 1
  freq[i] = ctr

# plt.figure(figsize=(15,7))
plt.bar( list(df.Type.unique()), list(freq.values()), color='aqua', width=0.6 )
plt.show()
```
